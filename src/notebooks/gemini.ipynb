```python
import time
from datetime import datetime, timedelta

# ‚öôÔ∏è CONFIGURACI√ìN: L√≠mite de tiempo m√°ximo (en minutos)
MAX_RESEARCH_TIME_MINUTES = 6  # Ajusta seg√∫n tu presupuesto (recomendado: 5-10 min)
MAX_RESEARCH_TIME_SECONDS = MAX_RESEARCH_TIME_MINUTES * 60

# ‚è±Ô∏è Iniciar Deep Research Agent (con b√∫squeda web autom√°tica)
print(f"üîç Iniciando Deep Research Agent...")
print(f"‚è∞ L√≠mite de tiempo: {MAX_RESEARCH_TIME_MINUTES} minutos")
start_time = datetime.now()
print(f"üïê Inicio: {start_time.strftime('%H:%M:%S')}\n")

interaction = client.interactions.create(
    input=prompt,
    agent='deep-research-pro-preview-12-2025',  # ‚Üê Deep Research Agent con b√∫squeda web autom√°tica
    background=True  # ‚Üê Obligatorio para tareas largas
)

interaction_id = interaction.id
print(f"‚úÖ Investigaci√≥n iniciada. ID: {interaction_id}\n")

# üìä Polling con l√≠mite de tiempo
elapsed_seconds = 0
poll_interval = 10  # Verificar cada 10 segundos
last_status = None

while elapsed_seconds < MAX_RESEARCH_TIME_SECONDS:
    try:
        # Obtener estado actual
        interaction = client.interactions.get(interaction_id)
        current_status = interaction.status

        # Solo imprimir si cambi√≥ el estado
        if current_status != last_status:
            elapsed_minutes = elapsed_seconds // 60
            elapsed_secs = elapsed_seconds % 60
            print(f"[{elapsed_minutes:02d}:{elapsed_secs:02d}] Estado: {current_status}")
            last_status = current_status

        # ‚úÖ Completado
        if current_status == "completed":
            print(f"\n‚úÖ Investigaci√≥n completada en {elapsed_seconds // 60} min {elapsed_seconds % 60} seg")

            # üí∞ Mostrar estad√≠sticas de uso (si est√°n disponibles)
            try:
                if hasattr(interaction, 'usage_metadata') or hasattr(interaction, 'usage'):
                    usage = getattr(interaction, 'usage_metadata', None) or getattr(interaction, 'usage', None)
                    if usage:
                        input_tokens = getattr(usage, 'prompt_token_count', None) or getattr(usage, 'input_token_count', None)
                        output_tokens = getattr(usage, 'candidates_token_count', None) or getattr(usage, 'output_token_count', None)
                        if input_tokens or output_tokens:
                            print("\n" + "="*80)
                            print("üí∞ ESTAD√çSTICAS DE USO:")
                            print("="*80)
                            if input_tokens:
                                print(f"üì• Tokens de entrada: {input_tokens:,}")
                            if output_tokens:
                                print(f"üì§ Tokens de salida: {output_tokens:,}")

                            # Estimaci√≥n de costo (precios aproximados para Gemini 3 Pro)
                            # Entrada: $2.00 por mill√≥n (hasta 200k), $4.00 por mill√≥n (>200k)
                            # Salida: $12.00 por mill√≥n (hasta 200k), $18.00 por mill√≥n (>200k)
                            cost_estimate = 0
                            if input_tokens:
                                if input_tokens <= 200000:
                                    cost_input = (input_tokens / 1_000_000) * 2.00
                                else:
                                    cost_input = (200000 / 1_000_000) * 2.00 + ((input_tokens - 200000) / 1_000_000) * 4.00
                                cost_estimate += cost_input

                            if output_tokens:
                                if output_tokens <= 200000:
                                    cost_output = (output_tokens / 1_000_000) * 12.00
                                else:
                                    cost_output = (200000 / 1_000_000) * 12.00 + ((output_tokens - 200000) / 1_000_000) * 18.00
                                cost_estimate += cost_output

                            if cost_estimate > 0:
                                print(f"üíµ Costo estimado: ${cost_estimate:.4f} USD")
                                print("‚ö†Ô∏è  Nota: Esta es una estimaci√≥n. Verifica el consumo real en Google Cloud Console.")
                            print("="*80)
            except Exception as e:
                # Si no se puede obtener el uso, no es cr√≠tico
                pass

            text_output = next((o for o in interaction.outputs if o.type == 'text'), None)
            if text_output:
                print("\n" + "="*80)
                print("üìÑ RESULTADO:")
                print("="*80 + "\n")
                print(text_output.text)
            break

        # ‚ùå Fall√≥
        elif current_status == "failed":
            print(f"\n‚ùå Investigaci√≥n fall√≥: {interaction.error if hasattr(interaction, 'error') else 'Error desconocido'}")
            break

        # ‚è≥ En progreso
        elif current_status == "in_progress":
            # Esperar antes de volver a verificar
            time.sleep(poll_interval)
            elapsed_seconds += poll_interval

        else:
            # Estado desconocido
            time.sleep(poll_interval)
            elapsed_seconds += poll_interval

    except Exception as e:
        print(f"\n‚ö†Ô∏è Error al verificar estado: {e}")
        print("Reintentando en 5 segundos...")
        time.sleep(5)
        elapsed_seconds += 5

# ‚è∞ Tiempo l√≠mite alcanzado
if elapsed_seconds >= MAX_RESEARCH_TIME_SECONDS and interaction.status != "completed":
    print(f"\n‚è∞ ‚ö†Ô∏è L√çMITE DE TIEMPO ALCANZADO ({MAX_RESEARCH_TIME_MINUTES} minutos)")
    print(f"üí° La investigaci√≥n puede continuar en segundo plano.")
    print(f"üìã Puedes recuperarla despu√©s usando:")
    print(f"   interaction = client.interactions.get('{interaction_id}')")
    print(f"\nüÜî Interaction ID: {interaction_id}")

    # Intentar obtener resultado parcial si existe
    try:
        current_interaction = client.interactions.get(interaction_id)
        if current_interaction.outputs:
            text_output = next((o for o in current_interaction.outputs if o.type == 'text'), None)
            if text_output:
                print("\nüìÑ Resultado parcial disponible:\n")
                print(text_output.text)
    except:
        pass
```
